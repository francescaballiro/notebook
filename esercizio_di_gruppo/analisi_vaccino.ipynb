```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('soggetti-guariti.csv')
```

```python
df.head(5)
```

```python
totali = df[['guariti_senza_somm', 'guariti_post_somm', 'guariti_post_1booster', 'guariti_post_2booster']].sum()
```

```python
print(totali)
```

```python
slices = totali.values
labels = totali.index
plt.pie(slices, labels = labels,
       autopct='%1.1f%%',
       wedgeprops={'edgecolor': 'black', 'linewidth': 1.5})
plt.title('Distribuzione dei guariti con o senza somministrazione vaccino')
plt.show()
```
